```python
import pandas as pd
test_data=pd.read_csv("../dataset/processed_testData.csv")
train_data=pd.read_csv("../dataset/processed_trainingData.csv")
```

```python
train_data = pd.concat([train_data.iloc[:, 0], train_data.iloc[:, -2], train_data.iloc[:, -1]], axis=1)
train_data
```

```python
import re
# 提取資訊
def extract_information(text):
    patterns = {
        'core_responsibilities': r'core responsibility (.*?) required skill',
        'required_skills': r'required skill (.*?) educational requirement',
        'educational_requirements': r'educational requirement (.*?) experience level',
        'experience_level': r'experience level (.*?) preferred qualification',
        'preferred_qualifications': r'preferred qualification (.*?) compensation benefit',
        'compensation_benefits': r'compensation benefit (.*)'
    }
    
    extracted_info = {}
    
    for key, pattern in patterns.items():
        match = re.search(pattern, text, re.IGNORECASE)
        extracted_info[key] = match.group(1).strip() if match else ''
    
    return pd.Series(extracted_info)

train_data[['core_responsibilities', 'required_skills', 'educational_requirements', 'experience_level', 'preferred_qualifications', 'compensation_benefits']] = train_data['Processed_model_response'].apply(extract_information)

train_data
```

```python
test_data = pd.concat([test_data.iloc[:, 0], test_data.iloc[:, -2], test_data.iloc[:, -1]], axis=1)
test_data = test_data[test_data['Processed_Resume_str'].apply(lambda x: isinstance(x, str))]
test_data
```

```python
# 提取資訊
def extract_resume_information(text):
    patterns = {
        'summary': r'summary (.*?) experience',
        'experience': r'experience (.*?) education',
        'education': r'education (.*?) skills',
        'skills': r'skills (.*)',
    }
    
    extracted_info = {}
    
    for key, pattern in patterns.items():
        match = re.search(pattern, text, re.IGNORECASE | re.DOTALL)
        extracted_info[key] = match.group(1).strip() if match else ''
    
    return pd.Series(extracted_info)

test_data[['summary', 'experience', 'education', 'skills']] = test_data['Processed_Resume_str'].apply(extract_resume_information)

# 輸出結果
# test_data=pd.concat([test_data.iloc[:, 0], test_data.iloc[:, 1], test_data.iloc[:, 3],test_data.iloc[:, 4],test_data.iloc[:, 5],test_data.iloc[:, 6]], axis=1)
test_data
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
import nltk
from nltk.corpus import stopwords
```

```python
# 下載停用詞列表
# nltk.download('stopwords')

# 加載停用詞列表
stop_words = set(stopwords.words('english'))

def preprocess_text(text):
    # 去除特殊字符和標點符號
    text = re.sub(r'\W', ' ', text)
    # 轉換小寫
    text = text.lower()
    # 去除多餘的空白字符
    text = re.sub(r'\s+', ' ', text)
    # 去除停用詞
    text = ' '.join(word for word in text.split() if word not in stop_words)
    return text

train_data['label'] = train_data['company_name'] + ' - ' + train_data['Processed_position_title']

train_data['Processed_model_response'] = train_data['Processed_model_response'].apply(preprocess_text)
test_data['Processed_Resume_str'] = test_data['Processed_Resume_str'].apply(preprocess_text)

# train_data['Processed_position_title'] = train_data['Processed_position_title'].apply(preprocess_text)
# test_data['Category'] = test_data['Category'].apply(preprocess_text)
```

```python
import os
import requests
import zipfile

# 下載模型
def download_glove_model(target_dir):
    
    if not os.path.exists(target_dir):
        os.makedirs(target_dir)
    
    # GloVe 4 模型下載連結
    url = "https://nlp.stanford.edu/data/glove.6B.zip"
    
    # 下載模型文件
    response = requests.get(url)
    
    # 將文件寫入到本地
    with open(os.path.join(target_dir, "glove.6B.zip"), "wb") as f:
        f.write(response.content)
    
    print("Download completed.")

# 解壓縮模型文件
def unzip_model(zip_file, target_dir):
    with zipfile.ZipFile(zip_file, 'r') as zip_ref:
        zip_ref.extractall(target_dir)

    print("Extraction completed.")


target_dir = "./glove_model"
download_glove_model(target_dir)
unzip_model(os.path.join(target_dir, "glove.6B.zip"), target_dir)
```

```python
from gensim.models import KeyedVectors
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

# GloVe 詞向量文件
glove_file = "./glove_model/glove.6B.300d.txt"

# 載入 GloVe 模型
model = KeyedVectors.load_word2vec_format(glove_file, binary=False, no_header=True)
model.wv=model
print("Model vocabulary:", len(model.wv.index_to_key))
```

```python
# 計算 GloVe 向量
def get_glove_vector(text, model):
    words = text.split()
    word_vectors = [model[word] for word in words if word in model.key_to_index]
    if word_vectors:
        return np.mean(word_vectors, axis=0)
    else:
        return np.zeros(model.vector_size)

job_vectors = np.array([get_glove_vector(text, model) for text in train_data['Processed_model_response']])
resume_vectors = np.array([get_glove_vector(text, model) for text in test_data['Processed_Resume_str']])

cosine_similarities = cosine_similarity(job_vectors, resume_vectors)

similarity_df = pd.DataFrame(cosine_similarities, index=train_data['company_name'], columns=test_data['ID'])
print(similarity_df)
```

```python
similarity_df.to_csv("resume_score.csv",index=False)
```

```python
def find_best_matches(similarity_df, top_n):
    best_matches = {}
    for job_idx in range(similarity_df.shape[0]):
        job = similarity_df.index[job_idx]
        sorted_candidates = similarity_df.iloc[job_idx].sort_values(ascending=False)
        top_matches = sorted_candidates.head(top_n)
        best_matches[job] = list(zip(top_matches.index.tolist(), top_matches.values.tolist()))
    return best_matches


best_matches = find_best_matches(similarity_df, 3)
output_df = pd.DataFrame(best_matches).T
output_df.columns = [f"Top_{i+1}_Resume" for i in range(3)]

```

```python
output_df['Top_1_Resume_ID'] = output_df['Top_1_Resume'].apply(lambda x: x[0])
output_df['Top_1_Similarity'] = output_df['Top_1_Resume'].apply(lambda x: x[1])

output_df['Top_2_Resume_ID'] = output_df['Top_2_Resume'].apply(lambda x: x[0])
output_df['Top_2_Similarity'] = output_df['Top_2_Resume'].apply(lambda x: x[1])

output_df['Top_3_Resume_ID'] = output_df['Top_3_Resume'].apply(lambda x: x[0])
output_df['Top_3_Similarity'] = output_df['Top_3_Resume'].apply(lambda x: x[1])

output_df.drop(columns=['Top_1_Resume', 'Top_2_Resume', 'Top_3_Resume'], inplace=True)

output_df.insert(0, 'Position', train_data['Processed_position_title'].values)

id_to_category = dict(zip(test_data['ID'], test_data['Category']))

output_df.insert(2, 'Top_1_Category', output_df['Top_1_Resume_ID'].map(id_to_category))
output_df.insert(4, 'Top_2_Category', output_df['Top_2_Resume_ID'].map(id_to_category))
output_df.insert(6, 'Top_3_Category', output_df['Top_3_Resume_ID'].map(id_to_category))
```
